L'exercici consisteix a crear un programa que et classifiqui una variable numèrica en funció de l’escala Suspès/Aprovat/Notable/Excel·lent. 

```python
x= float (input("Introdueix una nota: "))
if x<0:
    print ("Error, la nota introduida no es valida")
elif x>10:
    print ("Error, la nota introduida no es valida")
elif x<5:
    print ("Suspès")
elif x<7 :
    print ("Aprovat")
elif x<9:
    print ("Notable")
else:
    print ("Excel·lent")
```

- Exercici 4

Crea un programa que donada una llista qualsevol, et digui si es simètrica o no. Si ho és, que et digui quants elements té.

```python
x= int (input("Introdueix cuants elements te la teva llista: ")) 
l= []
for i in range (x):
    y= int (input("Introdueix el seguent element: ")) 
    l.append(y)
print (l)
z=0
t=len(l)-1
while (l[z]==l[t]and(z<t)):
    z+=1
    t-=1
if l[z]!=l[t]:
    print ("La llista no es simetrica")
else:
    print ("La llista es simetrica i té " +str(x)+" elements")
```

Crea un programa que donada una llista, et digui quants números coincideixen amb la seva posició. Per exemple [3,4,2,0,2,3,6] el 2 i el 6 coincideixen.

```python
x= int (input("Introdueix cuants elements te la teva llista: ")) 
l= []
for i in range (x):
    y= int (input("Introdueix el seguent element: ")) 
    l.append(y)
print (l)
z=0
for i in range (x):
    if l[i]==i:
        z+=1
print ("Hi han "+str(z)+" números que coincideixen amb la seva posició")
```
